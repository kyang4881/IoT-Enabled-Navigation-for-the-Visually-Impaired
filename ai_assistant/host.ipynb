```python
import os
import gradio as gr
from IPython.display import Audio, display
from gtts import gTTS
import time
import time
from Model import llm_model_interface
```

```python
llm=llm_model_interface.LLMInterface(
    file_name="./Data/brasbasah_cityhall_rafflesplace.pdf",
    model_name="anakin87/zephyr-7b-alpha-sharded",
    embedding_model_name="sentence-transformers/all-mpnet-base-v2"
)
```

```python
with gr.Blocks() as demo:
    chatbot = gr.Chatbot(label='MRT GPT')
    msg = gr.Textbox()
    clear = gr.ClearButton([msg, chatbot])
    msg.submit(llm.create_conversation, [msg, chatbot], [msg, chatbot])
demo.launch()
```

```python
chat_histories = llm.chat_histories
chat_histories_len = len(chat_histories)

while 1 == 1:
    new_chat_histories_len = len(chat_histories)
    if new_chat_histories_len != chat_histories_len:
        text = chat_histories[-1]
        tts = gTTS(text)
        tts.save("./output.wav")
        display(Audio("./output.wav", autoplay=True))
        chat_histories_len = len(chat_histories)
```
